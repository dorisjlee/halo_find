```python
m200b= halo_catalog[:,4]
```

```python
n, bins, patches  = plt.hist(np.log10(m200b),log=True,histtype='step')#, normed=True)
plt.xlabel("log10(m200b)",fontsize = 15)
plt.ylabel("Count",fontsize = 15)
```

```python
V = 10240.**3 # volume (divide by volume to get number density )
print V
```

```python
plt.semilogy(bins[:-1],n/V,'s')
plt.xlabel("log10(m200b)",fontsize = 15)
plt.ylabel("dN/dlnM ",fontsize = 15)
# plt.xticks(bins, ["10^%s" % i for i in bins])
```
