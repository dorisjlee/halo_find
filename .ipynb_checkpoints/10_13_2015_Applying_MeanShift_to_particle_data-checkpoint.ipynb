```python
prefix = "http://darksky.slac.stanford.edu/scivis2015/data/ds14_scivis_0128/"
ds = yt.load(prefix+"ds14_scivis_0128_e4_dt04_1.0000")
ad = ds.all_data()
x = ad[("all","particle_position_x")]
y = ad[("all","particle_position_y")]
z = ad[("all","particle_position_z")]
```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import seaborn as sns
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
n = 100000
ax.scatter(x[:n], y[:n], z[:n],',')#, c=c, marker=m)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```

```python
n = 50000
X= np.array([x[:n],y[:n],z[:n]])
X = X.T
```

http://scikit-learn.org/stable/modules/clustering.html#mean-shift

Advantage: don't have to specify bandwidth (radius to search through) or number of clusters 

n = 100000 takes quite a while to run ..

```python
from sklearn.cluster import MeanShift, estimate_bandwidth
bandwidth = estimate_bandwidth(X, quantile=0.2, n_samples=shape(X)[0])
ms = MeanShift(bandwidth=bandwidth, bin_seeding=True)
ms.fit(X)
labels = ms.labels_
centers = ms.cluster_centers_
labels_unique = np.unique(labels)
n_clusters_ = len(labels_unique)
```

```python
fig = plt.figure()
from mpl_toolkits.mplot3d import Axes3D
ax = fig.add_subplot(111, projection='3d')
color = ["red","green","blue","cyan","purple","orange","magenta","brown","yellow","lime"]
for i in range(n_clusters_):
    group  = np.where(labels==i)
#     plt.scatter(X[group,0], X[group,1],X[group,2],c=color[i])
    ax.scatter(X[group,0], X[group,1],X[group,2],c=color[i])#, c=c, marker=m)
plt.title('Estimated number of clusters: %d' % n_clusters_)
fig.show()
```
